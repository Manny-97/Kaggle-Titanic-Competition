## The RMS Titanic

The Royal Mail Steamer (RMS) Titanic was a British passenger liner that sank in the North Atlantic Ocean, 600 km southwest of Newfoundland in the early morning hours of April 15, 1912, after colliding with an icerberg at at around 23:40 (ship's time) on Sunday, April 14, 1912. She sank two hours and forty minutes later at 02:20 (ship's time; 05:18 GMT). There was an estimate of 2,224 passengers and crew aboard the ship, and more than 1500 people died, making it one of the deadliest commercial peacetime maritime disasters in modern history. The RMS Titanic was the largest ship afloat at the time it entered service and was the second of the three Olympic-class ocean liners operated by the White Star Line. The Titanic was built by the Harland and Wolff shipyard in Belfast.Thomas Andrews, her architect, died in the disaster.

Those who jumped off fell into the water and drowned or died within minutes due to effects of cold shock and incapacitation. The passengers' accommodation, especially the First Class section, was said to be "of unrivalled extent and magnificience", indicated by the fares that First Class accommodation commanded. Even Third Class, though considerably less luxurious than Second and First Classes, was unusually comfortable by contemporary standards.

Titanic's maiden voyage began shortly after noon on April 10, 1912 when she left Southampton on the first leg of her journey to New York. A few hours later she called at Cherbourg Harbour in northwestern France, a journey of 80 nautical miles (148km; 92miles) where she took on passengers. Her next port call was Queenstown (now Cobh) in Ireland, where she reached around midday on 11 April. She left in the afternoon after taking on more passengers and goods. By the time she departed westwards across the Atlantic she was carrying 892 crew members and 1,320 passengers.

The Titanic's high speed (22 knots) on water which ice had been reported was later criticised as reckless, but it reflected standard maritime practice at the time. The North Atlantic liners prioritised time-keeping above all other considerations. The shortage of lifeboats was not because of a lack of space nor because of cost. Now, removing outliers from both the Fare ad Age columns

```python
# So there are outliers in Fare, our approach is to remove those outliers.
for df in [train, test]:
    Q1 = df['Fare'].quantile(0.25)
    Q3 = df['Fare'].quantile(0.75)
    IQR = Q3 - Q1
    df['Fare'] = df['Fare'][~((df['Fare'] < (Q1 - 1.5*IQR)) | (df['Fare'] > (Q3 + 1.5*IQR)))]
```

```python
# So there are outliers in Age, our approach here as well isto remove such outliers
for df in [train, test]:
    Q1 = df['Age'].quantile(0.25)
    Q3 = df['Age'].quantile(0.75)
    IQR = Q3 - Q1
    df['Age'] = df['Age'][~((df['Age'] < (Q1 - 1.5*IQR)) | (df['Age'] > (Q3 + 1.5*IQR)))]
```

```python
# A naive GridSearch
best_score = 0
test_score = []
train_score = []
diff_accuracy = []
for criterion in ['entropy', 'gini']:
    for min_samples_leaf in [1, 2, 4, 5, 7, 10, 12, 15]:
        for min_samples_split in [2, 4, 5, 7, 10, 12, 15]:
            # For each combination of parameters, train an rfr
            rfr = RandomForestClassifier(criterion=criterion, min_samples_leaf=min_samples_leaf,\
                                        min_samples_split=min_samples_split, n_estimators=100, random_state=0)
            rfr.fit(X_train, y_train)
            score1 = rfr.score(X_train, y_train)
            # Evaluate rfr on the test set
            score = rfr.score(X_test, y_test)
            # If we got a better score, store the score and parameter
            train_score.append(score1)
            if score > best_score:
                test_score.append(score)
                best_score = score
                best_parameter = {'criterion': criterion, 'min_samples_leaf': min_samples_leaf,\
                                  'min_samples_split': min_samples_split}
                
print('The best accuracy is {:.2f}'.format(best_score*100))
print('The best parameter combination is {}'.format(best_parameter))
```

```python
# A simple performance reporting function
def clf_performance(classifier, model_name):
    print(model_name)
    print('Best Score:' + str(classifier.best_score_))
    print('Best Parameters:' + str(classifier.best_params_))
```
